```python
import pandas as pd
df=pd.read_csv("D:\\data science\\ML algorithms\\classification\\decision tree by hemanth\\salaries.csv")
```

```python
#df.head()
dir(df)
```

```python
inputs = df.drop('salary_more_then_100k',axis='columns')
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

```python
#machine learning algorithms understands only 1's and 0's(so we have to convert all the values into 0's and 1's)
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
```

```python
inputs.head()
```

```python
inputs_n=inputs.drop(['company','job','degree'],axis="columns")
```

```python
inputs_n.head()
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
#training the model
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target) #it will gives 1 because training data set and testing data set is same
```

```python
model.predict([[2,2,1]]) #google(2) sales executive(2) masters(1)
#answer will be no(0)
```

```python
model.predict([[2,0,1]]) #google(2) business manager(0) masters(1)
#answer will be yes(1)
```
