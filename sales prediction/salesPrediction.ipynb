```python
X = data.drop(columns="Sales")
Y = data["Sales"]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
train_data_prediction = model.predict(X_train)
train_mse = mean_squared_error(Y_train,train_data_prediction)
train_r2score = r2_score(Y_train,train_data_prediction)
print("Training data Metrics")
print("Mean Squared Error : ",train_mse)
print("r2_score : ",train_r2score)
```

```python
test_data_prediction = model.predict(X_test)
test_mse = mean_squared_error(Y_test,test_data_prediction)
test_r2score = r2_score(Y_test,test_data_prediction)
print("Training data Metrics")
print("Mean Squared Error : ",test_mse)
print("r2_score : ",test_r2score)
```
