```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits) #properties
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
digits.data[:2]
```

```python
df = pd.DataFrame(digits.data)
```

```python
df.head()
```

```python
#digits.target
df["target"] = digits.target
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'],axis = 'columns'),digits.target,test_size=0.2)
```

```python
#len(X_test)
#len(X_train)
#len(y_test)
#len(y_train)
#X_test
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier   #ensemble term used when ur using multiple algorithms
model = RandomForestClassifier(n_estimators=37)
model.fit(X_train,y_train)
#that's what we are doing here taking multiple decision trees
```

```python
model.score(X_test,y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
#better way to visualize the above one we use seaborn
```

```python
%matplotlib inline
import matplotlib. pyplot as pit
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot =True)
plt.xlabel("predicted")
plt.ylabel("truth")
```
